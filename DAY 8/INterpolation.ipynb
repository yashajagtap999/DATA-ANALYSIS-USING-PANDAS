```python
# interpolation:-- it is amethos used to replace the mising values with estimated value:--d1.interpolate()

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
d1={"name":["YAsh",None,"om","avi","prath"],
                 "salary":[100000,80000,None,60000,30000],
                 "performance":[100,90,None,30,40],
                 "age":[24,None,30,40,38]}
df=pd.DataFrame(d1)
print(df)

print("interpolated value:--")
df.interpolate(method='linear',axis=0,inplace=True)
print(df)
```
