```python
#to apply condition on columns:--
import pandas as pd
d1={"name":["Yash","omkar","om","shivam","yogesh"],
    "age":[21,22,23,24,25],
    "salary":[100000000,200000,30000,40000,90000]}
df=pd.DataFrame(d1)
d2=df[df["salary"]<100000]
print("filtered rows:--",d2)
```

```python
import pandas as pd
d1={"name":["Yash","omkar","om","shivam","yogesh"],
    "age":[21,22,23,24,25],
    "salary":[100000000,200000,30000,40000,90000]}
df=pd.DataFrame(d1)
print("Condition on the column")
d2=df[df["salary"]<100000]
print(d2)
```

```python
# apply multiple conditions:--
import pandas as pd
d1={"name":["Yash","omkar","om","shivam","yogesh"],
    "age":[21,22,23,24,25],
    "salary":[100000000,200000,30000,40000,90000]}
df=pd.DataFrame(d1)
print(df)
print("multiple conditions")

d2=df[(df["salary"]<100000)&(df['age']<25)]
print(d2)
```
