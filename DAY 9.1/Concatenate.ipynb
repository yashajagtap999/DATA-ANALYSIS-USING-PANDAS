```python
# concatenate:-- it is used to combine multiple dataframes together:-- pd.concat([d1,d2],axis=0/1,ignore_index=True)

import pandas as pd
d1=pd.DataFrame({"name":["Sundar","Elon","Yash"],
                 "age":[25,27,21]})

d2=pd.DataFrame({"name":["Ratan","Mukesh","Isha"],
                 "age":[80,59,19]})
data=pd.concat([d1,d2],axis=0,ignore_index=True)
print(data)
```
